```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

file_path = "final_clothes_data.xlsx"
df = pd.read_excel(file_path)
print("Dataset:")
print(df.head())
print()

def recommend_best_item(min_price, max_price, min_rating, df):
    best_items = []


    df['Price'] = df['Price'].replace('[\?,]', '', regex=True).astype(float)
    df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

    for index, row in df.iterrows():
        title = row['Title']
        rating = row['Rating']
        price = row['Price']
        if min_price <= price <= max_price and rating >= min_rating:
            best_items.append({'Title': title, 'Rating': rating, 'Price': price})

    best_items.sort(key=lambda x: x['Rating'], reverse=True)

    return best_items

min_price = 1000
max_price = 1200
min_rating = 4.4

recommended_items = recommend_best_item(min_price, max_price, min_rating, df)

if recommended_items:
    print("Recommended Items:")
    for item in recommended_items:
        print(item)
else:
    print("No items found within the specified price range and minimum rating.")
```
